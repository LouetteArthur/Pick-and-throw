```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File names
files = ['ppo_opt.csv', 'sac_opt.csv', 'td3_opt.csv','ppo_default.csv', 'sac_default.csv', 'td3_default.csv']

# Algorithms
algorithms = ['PPO', 'SAC', 'TD3 ']

# Hyperparameters
hyperparameters = ['Optuna' ,'Default']

# Variables to plot
variables = ['Reward', 'Time', 'Distance Ratio']

colors = [['red', 'orange'],['green', 'yellow'], ['blue', 'purple']]
colors_dict = {}
for i, algo in enumerate(algorithms):
    for j, par, in enumerate(hyperparameters):
            colors_dict[algo + ' ' + par] = colors[i][j]

for k, var in enumerate(variables):
    sns.set_style("darkgrid")
    fig, axs = plt.subplots(2, 3, figsize=(15, 15))
    for j, file in enumerate(files):
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file)
        # Create a distribution plot
        if j < 3:
            i = 0
        else:
            i = 1
            j = j%3
        
        sns.histplot(df[var], ax=axs[i, j], color= colors_dict[algorithms[j] + ' ' + hyperparameters[i]])

        # Set plot title and labels
        axs[i, j].set_title(algorithms[j] + ' ' + hyperparameters[i])

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np 
for file in files:
    df = pd.read_csv(file)
    print(file)
    print(f"{np.round(np.mean(df['Reward']), 3)} +/- {np.round(np.std(df['Reward']), 3)} & {np.round(np.mean(df['Success'])*100, 2)}\% & {np.round(np.mean(df['Time']), 3)} +/- {np.round(np.std(df['Time']), 3)} & {np.round(np.mean(df['Distance Ratio']), 2)} +/- {np.round(np.std(df['Distance Ratio']), 2)} ")
```
